```python
import pandas as pd 
import matplotlib.pyplot as plt

df = pd.read_excel('C:\\Users\\sour\\Downloads\\fin_zpl_2023.xlsx')
df_inf= pd.read_excel('C:\\Users\\sour\\Downloads\\inflation.xlsx')

sal_data=list(pd.Series(df.loc[int(0)][df.columns[1:].astype(int)]))
salb_data=list(pd.Series(df.loc[int(1)][df.columns[1:].astype(int)]))
salh_data=list(pd.Series(df.loc[int(2)][df.columns[1:].astype(int)]))
inf_data=list(df_inf[''].astype(float))
inf=inf_data[::-1]

def calculate_inf(a,b):
    c=[]
    for i in a:
        for k in b:
            if a.index(i)==b.index(k):
                c.append(i*(1-(k/100)))
    return c

y = pd.Series(df.loc[int(0)][df.columns[1:].astype(int)])
x = df.columns[1:].astype(int)

plt.plot(x, y, marker='.',label =' ')
plt.plot(x, calculate_inf(sal_data,inf), 'r', marker='.',label=' ')
plt.legend()
plt.title('          2000-2023 ')
plt.xlabel('')
plt.ylabel(' ')
plt.show()
```

```python
y_1 = pd.Series(df.loc[int(1)][df.columns[1:].astype(int)])

plt.plot(x, y_1, marker='.',label =' ')
plt.plot(x, calculate_inf(salb_data,inf), 'r', marker='.',label=' ')
plt.legend()

plt.title('          2000-2023 ')
plt.xlabel('')
plt.ylabel(' ')
plt.show()
```

```python
y_2 = pd.Series(df.loc[int(2)][df.columns[1:].astype(int)])

plt.plot(x, y_2, marker='.',label =' ')
plt.plot(x, calculate_inf(salh_data,inf), 'r', marker='.',label=' ')
plt.legend()

plt.title('             2000-2023 ')
plt.xlabel('')
plt.ylabel(' ')
plt.show()
```
